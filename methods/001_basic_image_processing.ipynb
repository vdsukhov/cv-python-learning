```python
import matplotlib.pyplot as plt
from skimage import io, color
from skimage.transform import rescale, resize, downscale_local_mean
```

```python
img = io.imread("../inpdata/images/Osteosarcoma_01.tif", as_gray=True)
print(img.shape, img.dtype)
```

```python
fig = plt.figure(figsize=(5, 5))
plt.imshow(img, cmap = "gray")
```

```python
img_rescaled = rescale(img, 0.25, anti_aliasing=False)
fig = plt.figure(figsize=(5, 5))
plt.imshow(img_rescaled, cmap = "gray")
```

```python
print("rescaled image shape:", img_rescaled.shape)
print("orig shape / 4:", tuple(map(lambda x: 0.25 * x, img.shape)))
```

```python
img_resized = resize(img, (200, 200), anti_aliasing=True)
fig = plt.figure(figsize=(5, 5))
plt.imshow(img_resized, cmap = "gray")
```

```python
img_downscaled = downscale_local_mean(img, (4, 3))
fig = plt.figure(figsize=(5, 5))
plt.imshow(img_downscaled, cmap = "gray")
```

```python
print(img_downscaled.shape)
```

```python
from skimage.filters import gaussian, sobel
```

```python
img = io.imread("../inpdata/images/Osteosarcoma_01_25Sigma_noise.tif")
fig = plt.figure(figsize=(10, 10))
plt.imshow(img, cmap = "gray")
```

```python
img_gaussian = gaussian(img, mode = "constant", cval = 0)
fig = plt.figure(figsize=(10, 10))
plt.imshow(img_gaussian)
```

```python
img_gray = io.imread("../inpdata/images/Osteosarcoma_01.tif", as_gray=True)
sobel_img = sobel(img_gray)
fig = plt.figure(figsize=(10, 10))
plt.imshow(sobel_img, cmap = "gray")
```

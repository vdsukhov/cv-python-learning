Let's take a look on the following equation:
$$U = O + \alpha \cdot (O - B),$$
where $U$ - unsharped image, $O$ - original image, $B$ - blurred image, $\alpha$ - is amount (some coefficient)

```python
from skimage import io, img_as_float
from skimage.filters import unsharp_mask, gaussian
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = img_as_float(io.imread("../inpdata/images/sandstone_blur_2sigma.tif", as_gray = True))

fig1 = plt.figure(figsize=(7, 7))
plt.imshow(img, cmap = "gray")
```

```python
gimg = gaussian(img, sigma=2, mode="constant", cval=0.0)

img2 = 2.0 * (img - gimg)
img3 = img + img2

fig2 = plt.figure(figsize=(7, 7))
plt.imshow(img3, cmap = "gray")
```

```python
img3 = unsharp_mask(img, radius=3, amount=2)
fig = plt.figure(figsize=(7, 7))
plt.imshow(img3, cmap = "gray")
```
